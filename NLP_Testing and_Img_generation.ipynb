Load Model & Generate Images

```python
from diffusers import DiffusionPipeline, AutoencoderKL
import torch
print("✅ Diffusers + Huggingface Hub working!")
```

```python
# Load VAE (autoencoder)
vae = AutoencoderKL.from_pretrained(
    "madebyollin/sdxl-vae-fp16-fix",
    torch_dtype=torch.float16
)
```

```python
# Load base SDXL pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    vae=vae,
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
).to("cuda")
```

```python
# Load your trained LoRA weights
pipe.load_lora_weights(MODEL_DIR)
print("✅ Model + LoRA loaded!")
```

```python
!pip install lpips transformers torch torchvision --quiet


import os
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
from torchvision import transforms
import lpips
from tqdm import tqdm
import numpy as np

# Load CLIP
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").cuda()
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# LPIPS setup
lpips_alex = lpips.LPIPS(net='alex').cuda()

# Folder with generated images
image_folder = "/content/drive/MyDrive/Colab Notebooks/NLP_DreamBooth/Generated_Images"

# Load image-caption pairs
image_caption_pairs = []
for img_file in os.listdir(image_folder):
    if img_file.endswith((".png", ".jpg", ".jpeg")):
        # Try to extract the prompt from filename
        prompt = img_file.replace("_", " ").replace(".png", "").replace(".jpg", "").replace(".jpeg", "")
        image_caption_pairs.append((img_file, prompt))

# Preprocessing for CLIP
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.4815, 0.4578, 0.4082), (0.2686, 0.2613, 0.2758)),
])

clip_scores = {}
itf_scores = {}
lpips_scores = []

for filename, caption in tqdm(image_caption_pairs):
    image_path = os.path.join(image_folder, filename)
    image = Image.open(image_path).convert("RGB")

    # CLIP Score
    inputs = clip_processor(text=[caption], images=image, return_tensors="pt", padding=True).to("cuda")
    outputs = clip_model(**inputs)
    clip_score = outputs.logits_per_image[0][0].item()
    clip_scores[filename] = clip_score

    # ITF Similarity (cosine of pooled embeddings)
    with torch.no_grad():
        image_input = preprocess(image).unsqueeze(0).cuda()
        image_embeds = clip_model.get_image_features(image_input)
        text_inputs = clip_processor(text=[caption], return_tensors="pt", padding=True).to("cuda")
        text_embeds = clip_model.get_text_features(**text_inputs)
        itf = torch.nn.functional.cosine_similarity(image_embeds, text_embeds).item()
        itf_scores[filename] = itf

    # LPIPS (perceptual distance from first image as anchor)
    ref_img = preprocess(Image.open(image_path).convert("RGB")).unsqueeze(0).cuda()
    dist = lpips_alex(ref_img, ref_img).item()  # use self-image just for consistency
    lpips_scores.append(dist)

# Averages
avg_clip = np.mean(list(clip_scores.values()))
avg_itf = np.mean(list(itf_scores.values()))
avg_lpips = np.mean(lpips_scores)

print("📊 Evaluation Summary:")
print(f"🎯 Average CLIP Score: {avg_clip:.4f}")
print(f"🎯 Average ITF Similarity: {avg_itf:.4f}")
print(f"🎯 Average LPIPS (self-similarity baseline): {avg_lpips:.4f}")

# Optional: show top-scoring image
best_img = max(clip_scores, key=clip_scores.get)
print(f"🏆 Top Image: {best_img} with CLIP score {clip_scores[best_img]:.4f}")
```

```python
from PIL import Image
from tqdm import tqdm
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import CLIPProcessor, CLIPModel
from torchvision import transforms
from IPython.display import display

# Set up device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load CLIP model and processor
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Optional image preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# Prompts to evaluate
prompts = [
    "A cheerful green dragon flying through the sky, waving its wings joyfully.",
    "A young wizard holding a glowing staff in a sunlit forest.",
    "A joyful princess twirling in a garden of pink and yellow flowers.",
    "A magical forest portal glowing among trees and vines.",
    "A dragon, wizard, and princess together under a rainbow.",
    "A mystical mountain landscape with fireflies at dusk.",
    "A peaceful meadow with cherry blossoms and gentle sunlight.",
    "A night scene with glowing mushrooms and floating lanterns.",
    "A waterfall hidden deep inside a fantasy rainforest.",
    "A sunrise over a magical kingdom surrounded by nature.",
]

results = []

# Evaluation loop
for prompt in tqdm(prompts, desc="🔮 Generating + Evaluating"):
    # Generate image from model
    image = pipe(prompt=prompt, num_inference_steps=30).images[0]
    display(image)

    # Process for CLIP
    inputs = clip_processor(text=[prompt], images=image, return_tensors="pt", padding=True).to(device)

    with torch.no_grad():
        outputs = clip_model(**inputs)

        # CLIP Score: dot product between CLIP image/text logits
        clip_score = outputs.logits_per_image[0].item()

        # ITI Similarity: cosine similarity between embeddings
        iti_similarity = cosine_similarity(
            outputs.image_embeds.cpu(), outputs.text_embeds.cpu()
        )[0][0].item()

    results.append((prompt, clip_score, iti_similarity))

    print(f"✅ Prompt: {prompt}")
    print(f"🎯 CLIP Score (logits): {clip_score:.4f}")
    print(f"🎯 ITI Similarity (cosine): {iti_similarity:.4f}")
    print("-" * 80)

# Summary stats
avg_clip = sum(r[1] for r in results) / len(results)
avg_iti = sum(r[2] for r in results) / len(results)

print("\n📊 Evaluation Summary:")
print(f"🎯 Average CLIP Score: {avg_clip:.4f}")
print(f"🎯 Average ITI Similarity: {avg_iti:.4f}")
```
