```python
from diffusers import DiffusionPipeline, AutoencoderKL
import torch
```

```python
vae = AutoencoderKL.from_pretrained(
    "madebyollin/sdxl-vae-fp16-fix",
    torch_dtype=torch.float16
)
```

```python
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    vae=vae,
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)
pipe.to("cuda")
```

```python
from PIL import Image
from sklearn.metrics.pairwise import cosine_similarity
import torch
import numpy as np
from transformers import CLIPProcessor, CLIPModel

clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to("cuda")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

image_folder = "/content/drive/MyDrive/Colab Notebooks/NLP_DreamBooth/Generated_Images"

# Modify to match your image names
captions = [
    "A young wizard with curly reddish-brown hair, casting a glowing spell in a storybook forest.",
    "A young wizard flying on a broomstick with a glowing wand, over hills and clouds.",
    "A young wizard reading a spellbook while holding a glowing wand in a dim castle.",
    "A young wizard standing in neutral pose with a glowing staff in warm watercolor tones.",
    "A young wizard walking through a forest path, casting a golden spell in hand-drawn style."
]

image_names = [
    "wizard_casting_spell.png", "wizard_generated_1.png", "wizard_generated_2.png",
    "wizard_generated_3.png", "wizard_generated_4.png", "wizard_generated_5.png"
]

clip_scores = []
missing = []

for i, (caption, image_name) in enumerate(zip(captions, image_names)):
    img_path = os.path.join(image_folder, image_name)

    if not os.path.exists(img_path):
        missing.append(f"❌ Missing: {image_name}")
        continue

    image = Image.open(img_path).convert("RGB")
    inputs = clip_processor(text=[caption], images=image, return_tensors="pt", padding=True).to("cuda")

    with torch.no_grad():
        outputs = clip_model(**inputs)
        score = cosine_similarity(outputs.image_embeds.cpu().numpy(), outputs.text_embeds.cpu().numpy())[0][0]
        clip_scores.append(score)
        print(f"✅ {image_name} → Score: {round(score, 4)}")

# Final Output
if clip_scores:
    print("\n🎯 Average CLIP Score:", round(np.mean(clip_scores), 4))
else:
    print("\n⚠️ No valid images found for scoring.")

if missing:
    print("\n🚫 Missing images:")
    for msg in missing:
        print(msg)
```

```python
!pip install lpips transformers torch torchvision --quiet

import os
import torch
import numpy as np
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
from lpips import LPIPS
from sklearn.metrics.pairwise import cosine_similarity
from transformers import AutoProcessor, AutoModel

# Paths
image_folder = "/content/drive/MyDrive/Colab Notebooks/NLP_DreamBooth/Generated_Images"
captions = {
    "wizard_casting_spell.png": "A young wizard with curly reddish-brown hair, casting a glowing spell in a storybook forest.",
    "wizard_generated_1.png": "A young wizard with a glowing staff standing in a magical forest in storybook style.",
    "wizard_generated_2.png": "A wizard with curly hair casting spells in a storybook castle.",
    "wizard_generated_3.png": "A storybook wizard in a blue robe holding a glowing wand, surrounded by sparkles.",
    "wizard_generated_4.png": "A wizard flying on a broomstick over a watercolor fantasy landscape.",
    "wizard_generated_5.png": "A wizard walking with a glowing spiral spell in a storybook forest."
}

# LPIPS
lpips_model = LPIPS(net='alex')
transform_lpips = transforms.Compose([transforms.Resize((256, 256)), transforms.ToTensor()])

# CLIP-like model for ITF similarity
processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
model = AutoModel.from_pretrained("openai/clip-vit-base-patch32")

def compute_itf_similarity(image, text):
    inputs = processor(text=[text], images=image, return_tensors="pt", padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return cosine_similarity(outputs.image_embeds.cpu(), outputs.text_embeds.cpu())[0][0]

lpips_scores, itf_scores = [], []

for fname in tqdm(captions):
    img_path = os.path.join(image_folder, fname)
    image = Image.open(img_path).convert("RGB")

    img_tensor = transform_lpips(image).unsqueeze(0)
    ref_tensor = transform_lpips(Image.open(os.path.join(image_folder, "wizard_casting_spell.png")).convert("RGB")).unsqueeze(0)

    lpips_scores.append(lpips_model(img_tensor, ref_tensor).item())
    itf_scores.append(compute_itf_similarity(image, captions[fname]))

print("✅ LPIPS Scores:", dict(zip(captions, lpips_scores)))
print("🎯 Average LPIPS:", np.mean(lpips_scores))

print("✅ ITF Similarity Scores:", dict(zip(captions, itf_scores)))
print("🎯 Average ITF Similarity:", np.mean(itf_scores))
```
